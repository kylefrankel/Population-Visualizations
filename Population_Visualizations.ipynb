```python
import pandas as pd
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/xialongtc10/MGT4250/main/gapminder.csv', encoding='ISO-8859-1')  #Change the file name to your data
print("dimension of data: {}".format(data.shape))
data.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 10))
data1 = data.loc[data['country']=='Canada']
plt.plot(data1["year"], data1["lifeExp"], "b", label="Canada")
data1 = data.loc[data['country']=='Mexico']
plt.plot(data1["year"], data1["lifeExp"], "r", label="Mexico")
data1 = data.loc[data['country']=='United States']
plt.plot(data1["year"], data1["lifeExp"], "g", label="United States")
plt.legend()
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.show()
```

```python
import plotly.express as px
```

```python
px.line(data.loc[data['country'].isin(['Canada', 'Mexico', "United States"])], 
  x='year',
  y = 'lifeExp', 
  color='country', 
  width=950,
  height=600
)
```
